### Both loc and iloc are row-first, column-second. This is the opposite of what we do in native Python, which is column-first, row-second.

```python
reviews.iloc[0].head()
```

```python
reviews.iloc[:, 0].head()
```

```python
reviews.iloc[:3, 0]
```

```python
reviews.iloc[1:3, 0]
```

```python
reviews.iloc[[0, 1, 2], 0] # It's also possible to pass a list.
```

```python
reviews.iloc[-5:]
```

```python
reviews.loc[0, 'country'] # Label-Based Selection
```

```python
reviews.loc[:, ['taster_name', 'taster_twitter_handle', 'points']]
```

### iloc uses the Python stdlib indexing scheme, where the first element of the range is included and the last one excluded. So 0:10 will select entries 0,...,9. loc, meanwhile, indexes inclusively. So 0:10 will select entries 0,...,10.

```python
reviews.set_index("title")
```

```python
reviews.country == 'Italy'
```

```python
reviews.loc[reviews.country == 'Italy'].head()
```

```python
reviews.loc[(reviews.country == 'Italy') & (reviews.points >= 90)].head() # with AND operator
```

```python
reviews.loc[(reviews.country == 'Italy') | (reviews.points >= 90)].head() # with OR operator
```

```python
reviews.loc[reviews.country.isin(['Italy', 'France'])].head()
```

```python
reviews.loc[reviews.price.notnull()].head()
```
