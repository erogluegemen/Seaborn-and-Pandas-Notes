```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
tips
```

```python
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day", size="size") # palette "YlGnBu"
```

```python
sns.histplot(tips["tip"], kde=True, bins=15)
```

```python
sns.displot(tips["tip"], kde=True, bins=15)
```

```python
sns.barplot(x="sex", y="tip", data=tips, palette="YlGnBu")
```

```python
sns.boxplot(x="day", y="tip", data=tips, hue="sex")
```

```python
sns.stripplot(x="day", y="tip", data=tips, hue="sex", dodge=True)
```

```python
sns.jointplot(x="tip", y="total_bill", data=tips, kind="reg") # kind="kde" , shade=True, cmap="YLGnBu" / kind="hex"
```

```python
sns.pairplot(titanic.select_dtypes(["number"]), hue="pclass")
```
